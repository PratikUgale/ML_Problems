```python
#import libraries

#Analysis
import pandas as pd

#Natural language processing
import nltk

#reguler expression
import re

#web scraping
import bs4 as bs

import requests
import heapq
```

1. we will url  wiki page https://en.wikipedia.org/wiki/World_war which contain information about world war
2. with request.get we will handle the request 
3. with beautiful soap library , we will exctract contain from web page and with pretiffy() , will display the web contain
4. now, data is in raw form. it contain all html tag present in the web page which is not usefull for us.
5. next step is now data cleaning

```python
sourse=requests.get('https://en.wikipedia.org/wiki/World_war')
su=sourse.content
```

```python
soup = bs.BeautifulSoup(su,'html.parser')
```

```python
soup.prettify()
```

1. If you look into web raw data, all the data which we are looking for in <p> tag
2. with find_all we will exctract all text data inside from <p> tag

```python
text=""
for paragraph in soup.find_all('p'):
    text +=paragraph.text
print(text)
```

1. After exctracting the data from <p> tag , we got useful data for summarization which we are looking for.
2. Still the data is in noisy form it contain lot of stopwords , spaces, and other impurities which effects its weight.
3. we will use python re library for data cleaning

```python
#Preprocess data
text = re.sub(r'\[[0-9]*\]',' ',text)
text = re.sub(r'\s+',' ',text)
#text = re.sub(r'\[\]')
clean_text=text.lower()
clean_text = re.sub(r'\W',' ',clean_text)
clean_text=re.sub(r'[^\w]', ' ', clean_text)
clean_text=re.sub(r'[^\[a-z]]', ' ', clean_text)
clean_text = clean_text.replace("u'", " ")
clean_text = re.sub(r'\d',' ',clean_text)
clean_text = re.sub(r'\s+',' ',clean_text)
#clean_text=str(unicode(clean_text))
print(clean_text)
```

Now its look like , data is neat and clean . nw we can do text summarization
1. from text data , our dataset contain 52 sentences
2. lets remove stopwords from data using nltk.corpus.stopwords.words

```python
#Tokenize the data
sentence=nltk.sent_tokenize(text)
length=len(sentence)
length
```

```python
#stopword
stop_words = nltk.corpus.stopwords.words('english')
```

1. for text summarization , we have to provide score for each sentences for that we have to provide score for each word present in sentence.
2. we can give score to word using its number of occurance present in each sentence.
3. below code will give dictionary for each word and its occorunce in each sentence
4. from output its look like world & war word occurs many time 

```python
table={}
for data1 in nltk.word_tokenize(clean_text):
    if data1 not in stop_words:
        
        if data1 not in table.keys():
            table[data1] = 1
        else:
            table[data1] +=1
    

table
```

1. after we get number of occurance for each word ,we can use it directly . we will divide each other word occurance with max word present in dataset which assign weight to each word

```python
max_weight=max(table.values())

for weight in table.keys():
    table[weight]=table[weight]/max_weight

table
```

1. now we will provide score to each sentence
2. we will select only those sentence which has 25 or less word present in it, we can increase this limit but I guess it will also effect results , after taking  different number I found 25 word for sentence will give pretty results
3. to display top 10 sentece , we will use heapq library which sort the result and gives top 10 sentence

```python
score_run={}
for sent in sentence:
        for word in nltk.word_tokenize(sent.lower()):
            
            if word in table.keys():
                
                if len(sent.split(' ')) < 25:
                    if sent not in score_run.keys():
                        score_run[sent] = table[word]
                    else:
                        score_run[sent] += table[word]
```

```python
best_sentences = heapq.nlargest(10,score_run , key=score_run.get)

print('---------------------------------------------------------')
for sentence in best_sentences:
    print(sentence)
```

# Sumy library

Sumy is python library which mostly used for text summarization, it has various types of API used for summarization
we will look into 1 2 such API for text summarization LSAtextsummarizer and textranksummarizer

let's import the sumy library and require summarizer API which we are using for summarization

```python
# Sumy Library for summarization

import sumy
from sumy.parsers.html import HtmlParser
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer
from sumy.nlp.stemmers import Stemmer
from sumy.utils import get_stop_words
 
from sumy.summarizers.luhn import LuhnSummarizer
from sumy.summarizers.edmundson import EdmundsonSummarizer
from sumy.summarizers.lex_rank import LexRankSummarizer
from sumy.summarizers.text_rank import TextRankSummarizer
```

```python
htmlpass=HtmlParser.from_url(url, Tokenizer('english'))
summarizer=LsaSummarizer()
summarizer = LsaSummarizer(Stemmer('english'))
summarizer.stop_words = get_stop_words('english')
summary=summarizer(htmlpass.document,10)
print(summary)
```

```python
summarizer_3 = TextRankSummarizer()
summarizer_3 = TextRankSummarizer(Stemmer('english'))
summarizer_3.stop_words = get_stop_words('english')
summary_3 =summarizer_3(htmlpass.document,2)
for sentence in summary_3:
    print(sentence)
```
